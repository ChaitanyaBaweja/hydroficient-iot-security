````python
# Save to JSON file
filename = "sensor_data.json"

with open(filename, "w") as f:
    json.dump(readings, f, indent=2)

print(f"‚úÖ Saved {len(readings)} readings to {filename}")

# Show file size
import os
size = os.path.getsize(filename)
print(f"üìÅ File size: {size:,} bytes")
````
